```python
# Import dependencies
# option 1
import matplotlib.pyplot as plt

# option 2
# from matplotlib import pyplot

# check version
# matplotlib.__version__
```

```python
# set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June","July","Aug","Sept", "Oct", "Nov", "Dec"]
          
# set the y-axis to a list of floats as the total fare in US doallrs accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Using MATLAB
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# Using Object-Oriented Interface
# create the plot with ax.plt()

# Step 1: create a figure
fig, ax = plt.subplots()   # variables: fig and ax

#Step 2: Define the axes
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()-option 2
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()-option 3
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# plt.show()
# looks for all the active figure objects and opens a window that dispalys them if you have two or more sets of data to plot
# should nly be use once per session

# create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotate Charts Using the MATLAB Method

# create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label ="Boston")

# create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45.
plt.ylim(0,45)

#create a title.
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="blue", linewidth=2, label ="Boston")

# create lebels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45.
plt.ylim(0,45)

#create a title.
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()

# Add a grid.
plt.grid()

```

```python
# practice with Object-Oriented Approach

# create the plot with ax.plt(), marker = diamond, line color = green, linewidth = 4, label = "Boston"
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="d", color ="green", linewidth =4, label ="Boston")

#plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label ="Boston")

# create labels for the x and y axes.
# plt.xlabel("Date")
# plt.ylabel("Fare ($)")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set the y limit between 0 and 45.
# plt.ylim(0,45)
ax.set_ylim(0,50)

# create a title.
# plt.title("PyBer Fare by Month")
ax.set_title("PyBer Fare by Month")

# Add the legend
# plt.legend()
ax.legend()

# Add a grid.
# plt.grid()
ax.grid()
```

#  5.1.5 Create a Bar Chart

```python
# Create a vertical bar chart
plt.bar(x_axis, y_axis)
```

```python
# Annotate the bar chart

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create a Horizontal Bar Chart

plt.barh(x_axis, y_axis)
```

```python
# Switch the axis

# Create a Horizontal Bar Chart

plt.barh(y_axis, x_axis)
```

```python
# y-axis to ascending orders (Jan on the top)
# gca() : get current axes
# invert_yaxis()

plt.barh(x_axis,y_axis, color ="magenta", label="Boston")
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```
